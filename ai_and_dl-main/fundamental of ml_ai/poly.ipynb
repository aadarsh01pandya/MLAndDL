```python
dataset='https://raw.githubusercontent.com/redashu/Datasets/master/PositionSalaries_Data.csv'
df = pd.read_csv(dataset)
```

```python
feature=df.iloc[:,1:-1].values
label=df.iloc[:,2:].values
```

```python
feature
```

```python
plt.scatter(feature,label)
plt.show()
```

```python
poly_regs= PolynomialFeatures() 
model=LinearRegression()
```

```python
x_poly= poly_regs.fit_transform(feature)  
```

```python
x_poly
```

```python
train_model=model.fit(x_poly,label)
```

```python
predict=train_model.predict( poly_regs.fit_transform(feature))
```

```python
predict
```

```python
print(label)
print(predict)
```

```python
plt.scatter(label,predict)
plt.plot(label,model.predict(poly_regs.fit_transform(feature)), color="red")  

plt.show()
```
