```python
import numpy as np
import cv2 
import skimage as ski
import matplotlib.pyplot as plt
```

```python
image1=plt.imread('cat.jpg')
plt.imshow(image1)
plt.axis('off')
```

```python
img_gray=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

print(img_gray.shape)
plt.imshow(img_gray,cmap='gray')
```

```python
th,img_binary=cv2.threshold(img_gray, 80,255, cv2.THRESH_BINARY)
img_binary.shape
```

```python
plt.imshow(img_binary,cmap='gray')
```

```python
th,img_binary_rgb=cv2.threshold(image1, 80,255, cv2.THRESH_BINARY)
img_binary_rgb.shape
```

```python
plt.imshow(img_binary_rgb,cmap='gray')
```

```python
plt.imshow(img_binary[100:200,200:400],cmap='gray')
```

```python
x,y,w,h=(50,150,180,140)
plt.imshow(img_binary[y:y+h, x:x+w],cmap='gray')
```

```python
x1,y1,x2,y2=(50,150,180,140)
plt.imshow(img_binary[y1:y1+y2,x1:x2+x1],cmap='gray')
```

```python
x1,y1,x2,y2=(50,150,180,240)
plt.imshow(img_binary[y1:y2,x1:x2],cmap='gray')
```

```python
img_large=cv2.resize(image1,(1500,1500), interpolation=cv2.INTER_CUBIC)
plt.subplot(1,2,1)
plt.imshow(img_large)

img_small=cv2.resize(image1,(0,0),fx=0.23,fy=0.25, interpolation=cv2.INTER_CUBIC)
plt.subplot(1,2,2)
plt.imshow(img_small)
```

```python
img_rot=image1[::-1]
plt.imshow(img_rot)
```

```python
img_x=ski.transform.rotate(image1,-30)
plt.imshow(img_x)
```

```python
img_90= cv2.rotate(image1,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_90)
```

```python
plt.subplot(1,3,1)
plt.title('orignal')
plt.imshow(image1[:,:,])

plt.subplot(1,3,2)
plt.title('adding')
plt.imshow(cv2.add(image1[:,:,],40))

plt.subplot(1,3,3)
plt.title('subtract')
plt.imshow(cv2.subtract(image1[:,:,],40))
```

```python
plt.subplot(1,3,1)
plt.title('orignal')
plt.imshow(image1[:,:,])

plt.subplot(1,3,2)
plt.title('Multiply')
plt.imshow(cv2.multiply(image1[:,:,],2))

plt.subplot(1,3,3)
plt.title('divide')
plt.imshow(cv2.divide(image1[:,:,],2))
```

```python

plt.imshow(cv2.add(image1[:,:,],image1[:,:,]))
```

```python
img_cntbrt= cv2.convertScaleAbs(
    image1,alpha=1.25,beta=-123
)
plt.imshow(img_cntbrt[:,:,:])
```

```python
img_blur=cv2.medianBlur(image1,1)
plt.imshow(img_cntbrt[:,:,:])
```
